import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```python
df = pd.read_csv("retail_sales_dataset.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
df.describe()
```

```python
sns.histplot(df['Total Amount'], bins=30)
plt.show()
```

Sales Distribution
The distribution of total sales is uneven, indicating that some transactions contribute more significantly than others.
A few higher-value transactions suggest the presence of premium customers.

```python
sns.barplot(x='Gender', y='Total Amount', data=df)
plt.show()
```

```python
sns.histplot(df['Age'], bins=20)
plt.show()
```

```python
sns.countplot(x='Product Category', data=df)
plt.xticks(rotation=45)
plt.show()
```
